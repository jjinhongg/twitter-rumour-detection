```python
def convert_label(label):
    if label == "rumour":
        return 1
    elif label == "non-rumour":
        return 0
    else:
        raise Exception("label classes must be 'rumour' or 'non-rumour'")


def convert_label_to_rumour(label):
    if label == 1:
        return "rumour"
    elif label == 0:
        return "non-rumour"
    else:
        raise Exception("label classes must be 1 or 0")


def get_labels(label_path, sourceIds):
    with open(label_path) as f:
        labels = json.load(f)
    corresponding_labels = [labels[id] for id in sourceIds]
    numeric_labels = [convert_label(label) for label in corresponding_labels]

    return numeric_labels

# define a function to get the rumours and non-rumours using test_texts and get_labels from the previous cell
def get_rumours_and_non_rumours(test_texts, test_labels):
    rumours = []
    non_rumours = []
    for i in range(len(test_texts)):
        if test_labels[i] == 1:
            rumours.append(test_texts[i])
        else:
            non_rumours.append(test_texts[i])
    return rumours, non_rumours
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

# Define functions for stopwords, bigrams, trigrams and lemmatization
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
rumours_words = list(sent_to_words(rumours))
non_rumours_words = list(sent_to_words(non_rumours))
data_words = list(sent_to_words(test_texts))

print(rumours_words[:1])

# Build the bigram and trigram models
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

# See trigram example
print(trigram_mod[bigram_mod[data_words[0]]])

# covid_dict = corpora.Dictionary(test_texts)
```

```python
# Remove Stop Words
rumours_words_nostops = remove_stopwords(rumours_words)
non_rumours_words_nostops = remove_stopwords(non_rumours_words)

# Form Bigrams
rumours_words_bigrams = make_bigrams(rumours_words_nostops)
non_rumours_words_bigrams = make_bigrams(non_rumours_words_nostops)

# Initialize spacy 'en' model, keeping only tagger component (for efficiency)
# python3 -m spacy download en
nlp = spacy.load('en', disable=['parser', 'ner'])

# Do lemmatization keeping only noun, adj, vb, adv
rumours_lemmatized = lemmatization(rumours_words_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])
non_rumours_lemmatized = lemmatization(non_rumours_words_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])

# print(rumour_lemmatized[:1])
```

# Find the most representative document for each topic

```python
def format_topics_sentences(ldamodel, corpus, texts):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row in enumerate(ldamodel[corpus]):
        print(row[0])
        row = sorted(row[0], key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)


rumours_df_topic_sents_keywords = format_topics_sentences(ldamodel=lda_model_rumours, corpus=rumours_corpus, texts=rumours_index)
non_rumours_df_topic_sents_keywords = format_topics_sentences(ldamodel=lda_model_non_rumours, corpus=non_rumours_corpus, texts=non_rumours_index)
```

```python
# Group top 5 sentences under each topic
rumours_sent_topics_sorteddf = pd.DataFrame()

rumours_sent_topics_outdf_grpd = rumours_df_topic_sents_keywords.groupby('Dominant_Topic')

for i, grp in rumours_sent_topics_outdf_grpd:
    rumours_sent_topics_sorteddf = pd.concat([rumours_sent_topics_sorteddf, 
                                             grp.sort_values(['Perc_Contribution'], ascending=[0]).head(1)], 
                                            axis=0)

# Reset Index    
rumours_sent_topics_sorteddf.reset_index(drop=True, inplace=True)

# Format
rumours_sent_topics_sorteddf.columns = ['Topic_Num', "Topic_Perc_Contrib", "Keywords", "Text"]

# Show
rumours_sent_topics_sorteddf.sort_values(by='Text', ascending=1).head(10)
```
